```python
# WAP to make a to do list program
def add_task(tasks):
  """Adds a task to the to-do list."""
  task = input("Enter the task: ")
  tasks.append({"task": task, "completed": False})  # Store tasks as dictionaries
  print("Task added!")

def view_tasks(tasks):
  """Views all tasks in the to-do list."""
  if not tasks:
    print("Your to-do list is empty.")
  else:
    print("Your to-do list:")
    for i, task_data in enumerate(tasks):
      task = task_data["task"]
      completed = task_data["completed"]
      status = "[x]" if completed else "[ ]"  # Mark completed tasks with [x]
      print(f"{i + 1}. {status} {task}")

def remove_task(tasks):
  """Removes a task from the to-do list."""
  view_tasks(tasks)
  if tasks:
    try:
      task_index = int(input("Enter the number of the task to remove: ")) - 1
      if 0 <= task_index < len(tasks):
        removed_task = tasks.pop(task_index)
        print(f"Removed task: {removed_task['task']}")  # Access task name
      else:
        print("Invalid task number.")
    except ValueError:
      print("Invalid input. Please enter a number.")

def mark_completed(tasks):
  """Marks a task as completed."""
  view_tasks(tasks)
  if tasks:
    try:
      task_index = int(input("Enter the number of the task to mark as completed: ")) - 1
      if 0 <= task_index < len(tasks):
        tasks[task_index]["completed"] = True  # Update task status
        print("Task marked as completed!")
      else:
        print("Invalid task number.")
    except ValueError:
      print("Invalid input. Please enter a number.")

def main():
  """Main function to run the to-do list program."""
  tasks = []
  while True:
    print("\nChoose an action:")
    print("1. Add task")
    print("2. View tasks")
    print("3. Remove task")
    print("4. Mark task as Completed")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
      add_task(tasks)
    elif choice == "2":
      view_tasks(tasks)
    elif choice == "3":
      remove_task(tasks)
    elif choice == "4":
      mark_completed(tasks)
    elif choice == "5":
      break
    else:
      print("Invalid choice. Please try again.")

if __name__ == "__main__":
  main()
```
